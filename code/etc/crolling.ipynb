```python
# States
country_dropdown = Select(browser.find_element(By.XPATH, '/html/body/div[1]/div/div[1]/div/div[1]/span/div/div/form/div/div[7]/div[1]/div/div/div[2]/div/div/div/select'))



states = {}
states_dropdown = browser.find_element(By.XPATH, "/html/body/div[1]/div/div[1]/div/div[1]/span/div/div/form/div/div[7]/div[2]/div/div/div[2]/div/div/div/select")
for country in customer_country:
    country_dropdown.select_by_value(country)
    if states_dropdown.get_attribute("disabled") == "true":
        continue
    else:
        states[country] = browser.find_element(By.ID, "fe88241").text.split("\n")[1:]
    time.sleep(1)
```
