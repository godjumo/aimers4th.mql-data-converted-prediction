```python
train.loc[(train["expected_timeline"]=="less_than_3_months") | 
          (train["expected_timeline"]=="less than 3 months. customer not answered . to call back") |
          (train["expected_timeline"]=="less than 3 months- outdoor led requiment") |
          (train["expected_timeline"]=="3 months") |
          (train["expected_timeline"]=="less than 3 months ,meeting with the customer for the more details and tentative boq will ne 32 and 43") |
          (train["expected_timeline"]=="duplicate lead - il220100042906. less than 3 months"), "expected_timeline"] = "less than 3 months"
train.loc[train["expected_timeline"]=="3_months_~_6_months", "expected_timeline"] = "3 months ~ 6 months"
train.loc[train["expected_timeline"]== "more_than_a_year", "expected_timeline"] = "more than a year"
train.loc[(train["expected_timeline"]== "9_months_~_1_year") |
          (train["expected_timeline"]== "9 months - 1 year"), "expected_timeline"] = "9 months ~ 1 year"
train.loc[train["expected_timeline"]== "6_months_~_9_months", "expected_timeline"] = "6 months ~ 9 months"
train.loc[~train["expected_timeline"].isin(test["expected_timeline"].unique()), "expected_timeline"] = np.nan
# Train과 Test 모두 NaN이 절반 정도 차지하고 있고, 
# NaN 중 is_converted가 True인 경우가 존재하므로 NaN을 Unknown으로 채워줌
train["expected_timeline"].fillna("Unknown", inplace=True)
test["expected_timeline"].fillna("Unknown", inplace=True)
```
